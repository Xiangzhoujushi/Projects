```python

import pandas as pd
xl_file = pd.ExcelFile("2016Consumer.xlsx")
dfs = {s: xl_file.parse(s) 
          for s in xl_file.sheet_names}
df = dfs["Table_2"]
li= ['Food', 'Alcoholic beverages', 'Housing', 'Apparel and services', 'Transportation','Healthcare', 'Entertainment', 'Personal care products and services', 'Reading', 'Education', 'Tobacco products and smoking supplies','Miscellaneous', 'Cash contributions', 'Personal insurance and pensions']
df_expenditure = df.loc[:1153] # get the expenditures.
df_expenditure_top = pd.DataFrame(df_expenditure.loc[df["Item"]==li[0]])
# df_expenditure_top
for i in range(1,len(li)):
    df_expenditure_top= df_expenditure_top.append(df_expenditure.loc[df["Item"]==li[i]], ignore_index=True)
df_expenditure_top[['All\nconsumer\nunits','Lowest\n20\npercent','Second\n20\npercent','Third\n20\npercent','Fourth\n20\npercent','Highest\n20\npercent']] = df_expenditure_top[['All\nconsumer\nunits','Lowest\n20\npercent','Second\n20\npercent','Third\n20\npercent','Fourth\n20\npercent','Highest\n20\npercent']].apply(pd.to_numeric)

df_expenditure_top=df_expenditure_top.drop(columns=['All\nconsumer\nunits'])
# df_expenditure_top.rename(columns={})
df_expenditure_top["Total expenditure"]=df_expenditure_top.sum(axis=1)

# the old names of the column
li1 = df_expenditure_top.columns
li2 = [x.replace('\n',' ') for x in li1]

df_expenditure_top=df_expenditure_top.rename(columns=dict(zip(li1,li2)))

df_expenditure_top.to_csv('out1.csv',index=False)
#table storing the percentage

# df_expenditure_top
df_expenditure_string = df_expenditure_top.to_json(orient='records')
# df_expenditure_string
# table storing the
```
